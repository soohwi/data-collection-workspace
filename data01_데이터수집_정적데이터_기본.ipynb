   - url : 요청할 대상 URL 문자열 지정
   - params : [선택적]요청시 전달 할  Query문자열 지정(딕셔너리, 튜플, 리스트, 바이트열 가능)

### class 속성 이용하기: tag_and_property

```python
ul_reply = bs_obj.find('ul')

ul_ko = bs_obj.find('ul', {'class': 'ko'})  # ul 엘리먼트의 class 속성의 값이 'ko'일 때 가져온다.
# ul_ko = bs_obj.find('ul', class_='ko') # 위 문장과 동일 문법! class 뒤에 언더바(_) 필수 ! 
print(ul_ko)
```

#### 정리 비교

| 방식 | 문법 | 실무 선호도 |
| --- | --- | --- |
| `find('ul', {'class': 'ko'})` | 딕셔너리 | 낮음 (속성 여러 개일 때만) |
| `find('ul', class_='ko')` | 키워드 인자 | 중간 (find 쓸 땐 이걸 선호) |
| `select_one('ul.ko')` | CSS 선택자 | **높음 (가장 많이 씀)** |

**요약**: `find` 계열을 쓴다면 `class_='ko'`가 딕셔너리보다 낫고, 더 나아가면 아예 `select()`/`select_one()`으로 넘어가는 게 실무 트렌드예요. 지금 수업 진도가 `find`/`find_all` 단계라면 이후에 `select` 나올 가능성이 높아요 — 나오면 그때 비교해서 알려드릴게요.

```python
ul_sns = bs_obj.find('ul', attrs={'class': 'sns'})  # ul엘리먼트 class 속성의 값이 'sns'일 때 가져온다.
print(ul_sns)
```

#### 정리
방식	속성 있을 때	속성 없을 때	실무 선호도
tag['href']	값 반환	KeyError (에러남)	낮음
tag.attrs['href']	값 반환	KeyError (에러남)	낮음
tag.get('href')	값 반환	None (안전)	높음

요약: 값이 확실히 있다고 100% 보장될 때만 ['href']를 짧게 쓰고, 대량의 데이터를 순회하며 크롤링할 때는 무조건 .get()을 쓰는 게 실무 표준이에요. 크롤링은 예외 상황(속성 누락)이 항상 생기기 때문에, 안전성이 최우선이거든요.
