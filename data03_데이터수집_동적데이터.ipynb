# webdriver-manager 라이브러리
webdriver-manager는 셀레니움(Selenium)으로 웹 크롤링이나 자동화를 할 때 크롬드라이버(ChromeDriver) 같은 브라우저 드라이버를 직접 다운로드받지 않고 코드로 자동 관리해 주는 라이브러리입니다.  
과거에는 크롬 브라우저가 업데이트될 때마다 수동으로 설치된 버전과 맞는 chromedriver.exe를 새로 받아 폴더에 넣어줘야 하는 번거로움이 있었는데, 이를 해결해 주는 필수적인 패키지입니다.

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager


# 시간 대기 작업할때 필요함
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
print(dir(webdriver))
print(webdriver.__version__)
```

```python
# 옵션 설정
options = Options()
options.add_argument('--start-maximized') # 화면 최대화

# 실행 시 최신 ChromeDriver를 자동으로 다운로드 및 로드합니다.
service = Service(ChromeDriverManager().install())

# 드라이버 실행
driver = webdriver.Chrome(service=service, options=options)
```

# Naver에서 많이 본 목록보기

```python
import random
```

```python
# 옵션 설정
options = Options()
options.add_argument('--start-maximized') # 화면 최대화

# 실행 시 최신 ChromeDriver를 자동으로 다운로드 및 로드합니다.
service = Service(ChromeDriverManager().install())

# 드라이버 실행
driver = webdriver.Chrome(service=service, options=options)
```

```python
import openpyxl


#엑셀 파일 생성
wb = openpyxl.Workbook()

#Sheet 워크시트 삭제
wb.remove(wb['Sheet'])

url = 'https://news.naver.com/'
driver.get(url)

# 2.0초 ~ 5.0초 사이의 랜덤한 실수 생성
time.sleep(random.uniform(2, 5))
# time.sleep(5)


# 랭킹 메뉴를 가져온다.
div_rank_btn = driver.find_element(By.CSS_SELECTOR, 'body > section > header > div.Nlnb._float_lnb > div > div > div > div > div > ul > li:nth-child(8) > a')
print(div_rank_btn.text)

div_rank_btn.click()

for i in range(1,4):
  # div:nth-child('+str(i)+') -> str(i)로 변환해서 넣어야 됨 !
  rank_name = driver.find_element(By.CSS_SELECTOR, '#wrap > div.rankingnews._popularWelBase._persist > div.rankingnews_box_wrap._popularRanking > div > div:nth-child('+str(i)+') > a')
  print(rank_name.text)

  # 워크시트 생성(index=1을 설정하면 가장 최근 워크시트가 제일 앞으로 배치됨)
  ws = wb.create_sheet(rank_name.text, index=1)
  ws['A1']='번호'
  ws['B1']='뉴스'

  div_rank_list = driver.find_elements(By.CSS_SELECTOR, '#wrap > div.rankingnews._popularWelBase._persist > div.rankingnews_box_wrap._popularRanking > div > div:nth-child('+str(i)+') > ul > li')
  #print(len(div_rank_list))
  #print(dir(div_rank_list))
  #print(div_rank_list)
  for idx, element_rank in enumerate(div_rank_list, start=2):
    num = element_rank.find_element(By.CLASS_NAME, 'list_ranking_num')
    content = element_rank.find_element(By.CLASS_NAME, 'list_content')
    #print(num.text, content.text)
    ws['A'+str(idx)] = num.text
    ws['B'+str(idx)] = content.text


wb.save('./news.xlsx')
```

```python
# 종료
driver.quit()
```

### 시간창 대기 3가지
- 일반적으로 시간창 대기는 time.sleep , implicitly_wait , Explicit Waits 3가지 사용
 - time.sleep: 그냥 지정한 시간만큼 기다린 뒤에 작업을 진행.
 - implicitly_wait : 지정한 시간만큼 기다려주고 그 이후는 기다리지 않음.
 - Explicit Waits : 특정 상태가 될때까지 기다리고, 상태가 되면 바로 다음 실행.

# 새 탭 이동하기

```python
# 옵션 설정
options = Options()
options.add_argument('--start-maximized') # 화면 최대화

# 실행 시 최신 ChromeDriver를 자동으로 다운로드 및 로드합니다.
service = Service(ChromeDriverManager().install())

# 드라이버 실행
driver = webdriver.Chrome(service=service, options=options)
```

```python
url = 'https://www.naver.com/'
driver.get(url)


# 2.0초 ~ 5.0초 사이의 랜덤한 실수 생성
time.sleep(random.uniform(2, 5))

# 뉴스메뉴 클릭
news = driver.find_element(By.CSS_SELECTOR,'#shortcutArea > ul > li:nth-child(5) > a')
news.click()

WebDriverWait(driver, random.uniform(2, 5)).until(EC.number_of_windows_to_be(2))

# 현재 열린 모든 창(탭)의 개수 확인
window_count = len(driver.window_handles)
print(f"현재 열린 창의 개수: {window_count}개")

# focus를  뉴스가 열린 탭으로 이동을 시킨다.
# 4. 포커스를 가장 최근에 열린 새 창(-1번 인덱스)으로 이동
# 열려있는 탭에 따라 인덱스가 붙음. 0, 1, 2.... (-1은 제일 마지막(최근))
# driver.switch_to.window(driver.window_handles[-1])
driver.switch_to.window(driver.window_handles[0])

# print(driver.page_source)
# 현재 포커스가 있는 창의 제목 가져오기
current_title = driver.title
print(f"현재 창의 제목: {current_title}")

try:
  a = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'body > section > header > div.Nlnb._float_lnb > div > div > div > div > div > ul > li:nth-child(3) > a' )))
  print(a.text)
except Exception as e:
  print(dir(e))
  print(e)
```

```python
# 종료
driver.quit()
```
