```python
import requests
```

```python
url = ''
# key = ''
# url += ''

response = requests.get(url)
print(response)                 # <Response [200]>
print(dir(response))
print(type(response))           # requests.models.Response
print(response.json())
```

```python
print(response.text)
```

```python
print(type(response.json()))  #dict
response.json()
```

```python
# item 접근
w_data = response.json()['items']['item']
print(type(w_data))               # list
print((len(w_data)))              # 갯수

for row in range(len(w_data)):
  print('spot_nm: ', w_data[row]['spot_nm'])
  print('경도: ', w_data[row]['lo_crd'])
  print('위도: ', w_data[row]['la_crd'])
```

```python
result = []
data = response.json()
w_data = data['items']['item']
for row in range(len(w_data)):
  line = {}
  line['spot_nm'] = w_data[row]['spot_nm']
  line['lo_crd'] = w_data[row]['lo_crd']
  line['la_crd'] = w_data[row]['la_crd']
  result.append(line)
```

```python
result
```

```python
import json

print(dir(json))
```

```python
#  ensure_ascii=False : 유니코드로 변환되는 작업 방지 
# indent ="\t" : 들여쓰기를 탭으로 설정
with open('./cycle.json', 'w',  encoding='utf-8') as f:
   json.dump(result, f, indent='\t', ensure_ascii=False)              # dump(obj, fp)
```

```python
with open('./cycle.json', 'r' , encoding='utf-8') as f:
  data = json.load(f)
  print(data)
```

# json
json.load  : 파일 읽기    
json.dump  : 파일 저장    
json.loads : 문자열(str) -> 객체(list, dict, tuple...)    
json.dumps : 객체(list, dict, tuple...) -> 문자열(str)

```python
json_str = '[1,2,3,4]'
print(type(json_str))
data = json.loads(json_str)
print(type(data))
print(data)
```

```python
json.dumps(data)
```

```python
json_str = '{"a":1, "b":2, "c":3}'
print(type(json_str))

data = json.loads(json_str)
print(type(data))
```

```python
json.dumps(data)
```
